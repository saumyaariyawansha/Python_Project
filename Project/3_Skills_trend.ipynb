```python
#import libraries
import pandas as pd
import seaborn as sns
import ast
import matplotlib.pyplot as plt
#load data
df = pd.read_csv('/Users/saumyaariyawansha/Documents/Data Science/python_exercises/data_jobs.csv')
#data cleaning
df['job_posted_date'] = pd.to_datetime(df.job_posted_date)
df['job_skills'] = df['job_skills'].apply(lambda skill_list: ast.literal_eval(skill_list) if pd.notna(skill_list) else skill_list)
```

```python
df_DA_US = df[(df['job_title_short'] == 'Data Analyst') & (df['job_country'] == 'United States')].copy()
```

```python
df_DA_US['job_posted_month'] = df_DA_US['job_posted_date'].dt.month
```

```python
df_DA_US_explode = df_DA_US.explode('job_skills')
```

```python
df_US_DA_pivot = df_DA_US_explode.pivot_table(index='job_posted_month', columns='job_skills', aggfunc='size', fill_value=0)

df_US_DA_pivot.loc['Total'] = df_US_DA_pivot.sum()
```

```python
df_US_DA_pivot
```

```python
df_US_DA_pivot = df_US_DA_pivot[df_US_DA_pivot.loc['Total'].sort_values(ascending=False).index]
```

```python
df_US_DA_pivot
```

```python
df_US_DA_pivot = df_US_DA_pivot.drop('Total')
```

```python
DA_monthly_total = df_DA_US.groupby('job_posted_month').size()

DA_monthly_total
```

```python
df_DA_US_percent = df_US_DA_pivot.div(DA_monthly_total/100, axis=0)
```

```python
df_DA_US_percent
```

```python
df_DA_US_percent = df_DA_US_percent.reset_index()
df_DA_US_percent['job_posted_month_name'] = df_DA_US_percent['job_posted_month'].apply(lambda x: pd.to_datetime(x, format='%m').strftime('%b'))
```

```python
df_DA_US_percent = df_DA_US_percent.set_index('job_posted_month_name')
df_DA_US_percent = df_DA_US_percent.drop(columns='job_posted_month')

df_DA_US_percent
```

```python
df_plot = df_DA_US_percent.iloc[:, :5]
```

```python
sns.lineplot(data=df_plot, dashes=False, palette='tab10')
sns.set_theme(style='ticks')
sns.despine()

plt.title('Trending Top Skills for Data Analysts in US')
plt.ylabel('Likeihood in Job Postings')
plt.xlabel('2023')
plt.legend().remove()

from matplotlib.ticker import PercentFormatter
ax = plt.gca()
ax.yaxis.set_major_formatter(PercentFormatter(decimals=0))

for i in range(5):
    plt.text(5, df_plot.iloc[8, i], df_plot.columns[i])
         
#from adjustText import adjust_text

#texts = []
#for i in range(5):
    #texts.append(plt.text(11.5, df_plot.iloc[-2, i], df_plot.columns[i]))
    
#adjust_text(texts, arrowprops=dict(arrowstyle="->", color='gray',lw=1))
```
